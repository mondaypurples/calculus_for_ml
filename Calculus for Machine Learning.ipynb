This mathematical function approximates the underlying function that describes how the features (columns in a dataset) are related to the target attribute. In statistics, linear regression is a linear(graphically represented as a straight line) approach for modelling the relationship between a scalar response and one or more explanatory variables.

Calculus provides a framework for understanding how mathematical functions behave. Using Calculus we can

#### Slope and y-intercept

Slope:
- The m value controls a line's slope
- The slope describes the steepness of the line

Another way to think of slope is rate of change. The rate of change is how much the y axis changes for a specific change in the x axis.

${m} = \frac{change in y}{change in x} $ = $ \frac{y} {x} $
    
Here is the slope calculation between points (1,4) and (4,13), which are both points on the line ${y} = 3x + 1 $

${m} = \frac{13 - 4}{4 - 1} $ = $\frac{9}{3} $ = 3

NOTE!

- Try changing x1 and x2 and observe the rate of change. No matter what the coordinates are, the rate of change between them is always 3.

- Straight, linear functions have a constant slope throughout.

y-intercept:

- The b value controls a line's y-intercept
- The y-intercept is the value where the line intersects the y-axis

#### Math behind slope

The rate of change between any points on a line, will always equal the slope of the function.

${y1} = {mx1 + b} $

${y2} = {mx2 + b} $

${y1 - y2} = {mx1 - mx2 + b - b} $

${y1 - y2} = {m(x1 - x2)} $

$m = \frac{(y1 - y2)}{(x1 - x2)} $

Where y1 - y2 represents y and x1 - x2 represents x.

```python
def slope(x1,x2,y1,y2):
    """
    Function calculates the rate of change between any two coordinates
    """
    m = (y1-y2)/(x1-x2)
    return m

slope_one = slope(0,4,1,13)
slope_two = slope(5,-1,16,-2)
```

```python
print(slope_one)
print(slope_two)
```

Thus, with linear functions we can determine the slope of the function from the equation itself and the slope is constant throughout.

## Secant Lines

### The slope between 2 points on a curve

When we calculate the slope between 2 points on a curve, we're really calculating the slope between the line that intersects both of those points.

And that line that intersects 2 points on a curve is known as a secant line.

Let's walk through the calculation for the line connecting the points at the x values 0 and 1.

Our non-linear function: ${y} = -({x^2}) + 3x -1 $

x1 = 0

x2 = 1

We use ${y} = -({x^2}) + 3x -1 $ to calculate y1 and y2.

${y1} = -({x1^2}) + 3x1 -1 $ = 0 + 0 - 1 = -1

y1 = -1

${y2} = -({x2^2}) + 3x2 -1 $ = -1 + 3 - 1 = 1

y2 = 1

Slope:
    
$m = \frac{(y1 - y2)}{(x1 - x2)} $

m = 2/1 = 2

Because both (x1, y1) and (x2,y2) are points on the line, we can use either one to calculate y-intercept (b).

${y1} = {m}{x1} + {b} $

-1 = 2(0) + b

-1 = b

b = -1

So now we have m = 2, b = -1. So can we declare the equation for the line connecting both points:

$y = 2x -1 $
    

### Function to draw secant lines

```python
import seaborn
seaborn.set(style='darkgrid')

def draw_secant(x_values):
    """
    Function accepts 2 x-values as a list
    Draws a secant line connecting those 2 points.
    
    Function:
    Determines the y-values for those 2 x-values using our non-linear function
    Calculates the slope between those 2 points
    Calculates the y-intercept using arithmetic
    Plots the secant line using green color
    Displays all of the plots in the function
    """
    x = np.linspace(-20,30,100) # Generates a numpy array of 100 values from -20 to 30
    y = -1*(x**2) + x*3 - 1
    plt.plot(x,y)
    
    x_0 = x_values[0]
    x_1 = x_values[1]
    y_0 = -1*(x_0**2) + x_0*3 - 1
    y_1 = -1*(x_1**2) + x_1*3 - 1
    m = (y_1 - y_0) / (x_1 - x_0)
    b = y_1 - m*x_1
    
    y_secant = x*m + b
    plt.plot(x, y_secant, c='green')
    plt.show()
    
    
draw_secant([3,5])
draw_secant([3,10])
draw_secant([3,15])
draw_secant([3,25])
```

### Observation

In the first graph, draw_secant([3,5]) the gap between x0 and x1 was small. The secant line intersected comfortably with our curve.

As we increase the gap between the 2 x-values, the secant line started to diverge from our curve. This observation is critical to understanding how slope works for curves.

The slope of a curve at a specific point, x1, is best understood as the slope of the secant line at increasingly smaller intervals of [x0, x1]. The smaller the difference between x0 and x1, the more precise the secant line approximates the slope at that point on our curve.

### What happens when x0 and x1 are the same? 

```python
draw_secant([3,3])
```

When x0 and x1 are the same, the slope value of the curve is undefinded. This is because division by 0 has no meaning in mathematical calculations.

## Tangent Line

```python
draw_secant([3,2.9])
draw_secant([3,2.99])
draw_secant([3,2.999])
draw_secant([3,3.01])
draw_secant([3,3.001])
draw_secant([3,3.0001])
draw_secant([3,3.00001])
```

When we keep x1 = 3 and x0 very very close to 3 and not exactly 3 (to avoid division by 0), we can see the slope value appraches -3.

m = -3

This value is known as the instantaneous rate of change, because it describes the slope at a particular point.

For linear functions, the instantaneous rate of change at any point on the line is the same. For nonlinear functions, the instantaneous rate of change describes the steepness of the graph of the function at a specific point.

This line is known as the tangent line and it only intersects our function at one point. 
Secant line connects 2 points that are increasingly close together. Think of the tangent line as the secant line when both points are the same.


Now we know how slope works for linear and nonlinear functions!!

## Understanding Limits

Limits - A mathematical tool that will enable us to calculate the slope at any point along a nonlinear function.

To calculate the slope of the tangent line at a specific point using LIMITS!

We need to be able to state and reason about what value the slope approaches as x2 approaches x1, despite the divison by zero problem we encounter when x1 = x2.

How do we do that? To to that, we need to reframe the problem as a LIMIT.

A LIMIT describes the value a function approaches when the input variable (x2) to the function $m = \frac{f(x2) - f(x1)}{(x2 - x1)} $ approaches a specific value.



Mathematical Notation of this statement "As x2 approaches 3, the slope between x1 and x2 approaches -3" using LIMIT is as follows:



Our non-linear function: ${y} = -({x^2}) + 3x -1 $

${f(3)} = -({3^2}) + 3*3 -1 $

${f(3)} = -1 $

### Prove that this limit actually does equal 3

#### Defined vs. Undefined Limits

Undefined Limits:
The limit we have above is an undefined limit because plugging in 3 for x2 results in an undefined value.

Defined Limits:

Slope equation: 
    
$m = \frac{f(x1) - f(x0)}{(x1 - x0)} $

Our non-linear function: 

${y} = -({x^2}) + 3x -1 $

x1 = 2.9

x0 = 3

${f(2.9)} = -({2.9^2}) + {3*2.9} -1 $ 

${f(3)} = -({3^2}) + {3*3} -1 $ 

m = -0.71 - (-1) / (2.9 - 3)

m = 0.29 / -0.1 = - 2.9



$m = \frac{f(2.9) - f(3)}{2.9 - 3} $

$m = -2.9 $

Thus, by rewriting and converting an undefined limit to a defined limit, we can solve the limit using direct substitution and standard arithmetic.

## Compute limits using Python

The [SymPy](https://docs.sympy.org/dev/tutorial/calculus.html#limits) library has a suite of functions that let us calculate limits. 

```python
# import sympy
import sympy
# declare variables
x,y = sympy.symbols('x y')
y = x**2 + 3*x
print(type(x))
print(type(y))
```

Note that x, y point to a special SymPy object (not a list).

In SymPy, the workflow is different and the Python variables we use don't directly map to specific values. In SymPy, Python variables map directly to variables in math. We express a mathematical function as a transformation of the Python variable. When we called sympy.symbols() and passed in x, x points to a special SymPy object (not a list).

```python
# import sympy
import sympy

# declare variables
x2,y = sympy.symbols('x2 y')

# the function we're taking the limit for y = -(x ** 2) + 3x -1
# the input variable x2
# the value the input variable approaches 2.9

# calculate limit using sympy.limit() 
limit_one = sympy.limit((-x2 ** 2 + 3*x2-1+1) / (x2-3) , x2, 2.9 )
print(limit_one) # SymPy returned  2.9 as the slope of the secant line which matches our calculation using the slope formula
```

SymPy returned 2.9 as the slope of the secant line, which matches our calculation using the slope formula.

## Properties of Limits

### First property

```python
import sympy
x,y = sympy.symbols('x y')

# three parameters for sympy.limit()
# the function we're taking the limit for y = 3*(x**2) + 3*x - 3
# the input variable x
# the value the input variable approaches 1

y = 3*(x**2) + 3*x - 3
limit_two = sympy.limit(y, x, 1)
print(limit_two)
```

### Second Property

```python
import sympy
x,y = sympy.symbols('x y')

# three parameters for sympy.limit()
# the function we're taking the limit for y = x ** 3 + 2*(x**2) - 10*x
# the input variable x
# the value the input variable approaches -1

y = x ** 3 + 2*(x**2) - 10*x
limit_three = sympy.limit(y, x, -1)
print(limit_three)
```

## Undefined Limit To Defined Limit

Let's convert the original undefined limit that calculated the slope at x=3 into a defined limit:

Now we can use direct substitution to arrive at 3.

```python
import sympy
x2, y = sympy.symbols('x2 y')

# three parameters for sympy.limit()
# the function we're taking the limit for (-x2**2 +3*x2 - 1 + 1) / (x2 - 3)
# the input variable x2
# the value the input variable approaches 3

y = (-x2**2 +3*x2 - 1 + 1) / (x2 - 3)
limit_four= sympy.limit(y, x2, 3)
print(limit_four)
```

Thus, we have calculates the slope of the tangent line at a specific point using limits!!!

## Extreme Values

Critical points are interesting to us in data science when they represent extreme values.

Extreme values can be split into two categories -- minimum and maximum.

Minimum - a critical point can be a minimum value if the slope transitions from negative to positive at that point.

Maximum - if the slope transitions from positive to negative at a point

However, in the function $ y = x ^ 3 $ , at x = 0, the slope does not transition at a ctitical point. The slope remains positive before and after the critical point. In this case, it can't be a minimum or a maximum point.

## Example - Curve's three critical points

A curve can have multiple critical points and multiple candidates for minimum and maximum values. The following curve has 3 critical points and all 3 meet the criteria of transitioning slope from positive to negative or negative to positive. At the same time, none of the values are the highest points on the curve (while one of them seems to be the lowest).

These values are known as local extrema, since they're the lowest or highest value in the interval surrounding each point. 

If a critical point is the lowest point in the interval surrounding it, it's known as a relative minimum. 

If it's the highest point in the neighborhood, it's known as a relative maximum.



Now let us take the derivative function y= 2x +3 and set it to 0. Because, we just learned that relative minimums and maximums can occur when slope is 0.

#### How to build this sign chart?


In order to build this sign chart, use the derivative function: $ {3x^2} - 2x $ and values of x to the left and right of your critical points (ie the values for x which make the derivative collapse to 0). (ie x = 0.66). 

Once built, your sign chart shows you where your original function ($ {f(x)} = {x}^3 - {x}^2 $) has tangent lines with slopes that are either pos, neg, or zero. Knowing if the slope of the tangent is pos or neg to the left/right of our critical points tells us if that critical point is a relative min or max or neither.

The key point to keep straight in your head while working with these two functions is that plugging in values for x into the original function ($ {f(x)} = {x}^3 - {x}^2 $) will give you y values which then gives you a coordinate pair (x, y) that helps to generate a graph of the original function. Plugging in values for x into the derivative function ($ {3x^2} - 2x $) will give you the value of the slope of the tangent at that point in the original function.

For example, if we use x = 1 in the derivative function, we get a result of 1. This means that if you were to draw a tangent line to the graph x^3 - x^2 at x = 1, the slope of that tangent line would be 1 (ie a line at 45 degree angle). Similarly, at x = -1, the slope of the tangent to the graph x^3 - x^2 would be 5 (ie a line at ~78.7 degree angle).
